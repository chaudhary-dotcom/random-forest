```python
import numpy as np
import pandas as pd
from sklearn.ensemble import RandomForestClassifier, BaggingClassifier
from sklearn.tree import plot_tree
import matplotlib.pyplot as plt
from sklearn.datasets import make_classification
```

```python
X,y = make_classification(n_samples=100,n_features=5, n_redundant=0, n_informative=5, n_clusters_per_class=1)
```

```python
df = pd.DataFrame(X, columns=['col1', 'col2', 'col3', 'col4', 'col5'])
df['target'] = y
df.head()
df.shape
```

```python
bag = BaggingClassifier(max_features=3)
bag.fit(df.iloc[:,:3], df.iloc[:,-1])
plt.figure(figsize=(12,12))
plot_tree(bag.estimators_[0], filled=True)
```

```python
rf = RandomForestClassifier(max_features=3)
rf.fit(df.iloc[:,0:-1], df.iloc[:, -1])
plot_tree(rf.estimators_[4], filled=True)
```
